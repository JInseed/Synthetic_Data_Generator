# Generate Data_Synthpop

```python
# 변수 타입 지정, 'int', 'float', 'datetime', 'category', 'bool' 만 가능
my_dtypes = {"저소득층여부": "category",
 "나이": "int",
 "교육수준": "category",
 "혼인상태": "category",
 "가구형태": "category",
 "주택유형": "category",
 "주택점유형태": "category",
 "총생활비": "int",
 "총소득": "int",
 '가구서비스': 'category',
 '노인서비스': 'category',
 '아동서비스': 'category'
}

my_dtypes
```
