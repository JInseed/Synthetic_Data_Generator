# Generate Data_CTGAN

```python
# 범주형 변수 지정
categorical_features = [
    '저소득층여부',
    '교육수준',
    '혼인상태',
    '가구형태',
    '주택유형',
    '주택점유형태',
    '가구서비스',
    '노인서비스',
    '아동서비스'
]

ctgan = CTGAN(epochs = 150, batch_size = 100, verbose=True)
ctgan.fit(raw, categorical_features)

# Create synthetic data
nrow = len(raw)
synthetic_data = ctgan.sample(nrow)
```
